```python
import pandas as pd
import numpy as np
```

```python
#Combining monthly data at once
jan = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2101.csv")
feb = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2102.csv")
mar = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2103.csv")
apr = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2104.csv")
may = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2105.csv")
jun = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2106.csv")
jul = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2107.csv")
aug = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2108.csv")
sep = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2109.csv")
oct = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2110.csv")
nov = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2111.csv")
dec = pd.read_csv("/Users/nakyungkim/Desktop/flight/2021/2112.csv")
```

```python
flight_2019 = pd.concat([jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec], ignore_index=True)

flight_2019.head()
```

```python
flight_2019.drop(['DEP_DELAY_NEW'], axis=1, inplace=True)
flight_2019.head()
```

```python
# Check the data types
flight_2019.dtypes
```

```python
# Count the NaN values
flight_2019.isna().sum()
```

```python
# Drop the the row which has NaN values
flight_2019.dropna(subset=['DEP_TIME', 'DEP_DELAY'], inplace=True)
flight_2019.head()
```

```python
# Check the values again
flight_2019.isna().sum()
```

```python
# Make the copy file
flight_2019_copy = flight_2019.copy()
flight_2019_copy.head()
```

```python
# Filter the 4 biggest City in America
excluded_cities = ["New York, NY", "Los Angeles, CA", "Chicago, IL", "Houston, TX"]

# Choose the value from ORIGIN_CITY_NAME columns
filtered_data = flight_2019_copy[flight_2019_copy['ORIGIN_CITY_NAME'].isin(excluded_cities)]

# Check the data
filtered_data.head(5)
```

```python
# Check the row if it is filtered well
other_cities = filtered_data[~filtered_data['ORIGIN_CITY_NAME'].isin(excluded_cities)]

other_cities.head()
```

```python
# Filter the 8 Airlines in America
airline_code = ['DL', 'WN', 'UA', 'G4', 'AA', 'NK', 'F9', 'B6']

# Choose the data from OP_CARRIER columns
filtered_data_final = filtered_data[flight_2019_copy['OP_CARRIER'].isin(airline_code)]

filtered_data_final.head(10)
```

```python
# # Check the row if it is filtered well
other_airlines = filtered_data_final[~filtered_data['OP_CARRIER'].isin(airline_code)]

other_airlines.head()
```

```python


# FL_DATE 열의 데이터 타입을 datetime으로 변환합니다.
filtered_data_final['FL_DATE'] = pd.to_datetime(filtered_data_final['FL_DATE'].str.split().str[0])

# 월, 일, 연도를 새로운 열에 각각 넣습니다.
# filtered_data_final['Year'] = filtered_data_final['FL_DATE'].dt.year
# filtered_data_final['Month'] = filtered_data_final['FL_DATE'].dt.month
# filtered_data_final['Day'] = filtered_data_final['FL_DATE'].dt.day
```

-- 동혁이가 한 것

```python
#Change the format of the ORIGIN_CITY_NAME (New York, NY) -> New York
filtered_data_final['ORIGIN_CITY_NAME'] = filtered_data_final['ORIGIN_CITY_NAME'].str.replace(', \w\w$', '', regex=True)
filtered_data_final.tail(5)
```

```python
filtered_data_final['FLIGHT_DELAY_LEVEL'] = 0
filtered_data_final.loc[filtered_data_final['DEP_DELAY'] < 15, 'FLIGHT_DELAY_LEVEL'] = 0
filtered_data_final.loc[(filtered_data_final['DEP_DELAY'] >= 15) & (filtered_data_final['DEP_DELAY'] < 45), 'FLIGHT_DELAY_LEVEL'] = 1
filtered_data_final.loc[(filtered_data_final['DEP_DELAY'] >= 45) & (filtered_data_final['DEP_DELAY'] < 75), 'FLIGHT_DELAY_LEVEL'] = 2
filtered_data_final.loc[(filtered_data_final['DEP_DELAY'] >= 75) & (filtered_data_final['DEP_DELAY'] < 105), 'FLIGHT_DELAY_LEVEL'] = 3
filtered_data_final.loc[(filtered_data_final['DEP_DELAY'] >= 105) & (filtered_data_final['DEP_DELAY'] < 135), 'FLIGHT_DELAY_LEVEL'] = 4
filtered_data_final.loc[filtered_data_final['DEP_DELAY'] >= 135, 'FLIGHT_DELAY_LEVEL'] = 5
filtered_data_final.tail()
```

```python
num_unique_levels = filtered_data_final['FLIGHT_DELAY_LEVEL'].nunique()
print("FLIGHT_DELAY_LEVEL의 총 종류 수:", num_unique_levels)
```
